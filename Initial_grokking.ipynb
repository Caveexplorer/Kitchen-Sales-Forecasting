```python
from statsmodels.tsa.arima.model import ARIMA

# Assuming an 80-20 train-test split
train_size = int(len(model_fodder) * 0.8)
train, test = model_fodder[:train_size], model_fodder[train_size:]

#Setting date as index
train.set_index('DATE', inplace=True)
test.set_index('DATE', inplace=True)


# Build ARIMA model on training data
model = ARIMA(train['PRCP'], order=(0,0,0))
model_fit = model.fit()
```

```python
# Forecast values for the test set
forecast = model_fit.forecast(steps=len(test))

# Evaluation metrics like Root Mean Squared Error (RMSE)
from sklearn.metrics import mean_squared_error
from math import sqrt

rmse = sqrt(mean_squared_error(test['PRCP'], forecast))
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
from pmdarima import auto_arima
import pandas as pd

# Assuming model_fodder is your original DataFrame
train_size = int(len(model_fodder) * 0.8)
train, test = model_fodder[:train_size], model_fodder[train_size:]

# Convert the index to datetime for train and test
train.set_index('DATE', inplace=True)
train.index = pd.to_datetime(train.index)
test.set_index('DATE', inplace=True)
test.index = pd.to_datetime(test.index)

best_rmse = float('inf')
best_m = None
best_order = None
best_seasonal_order = None
best_model = None

# Loop through possible seasonal frequencies
for m in range(1, 13):
    temp_model = auto_arima(train['PRCP'], seasonal=True, m=m, trace=True, error_action='ignore', suppress_warnings=True)
    future_forecast = temp_model.predict(n_periods=len(test))
    rmse = sqrt(mean_squared_error(test['PRCP'], future_forecast))

    if rmse < best_rmse:
        best_rmse = rmse
        best_m = m
        best_order = temp_model.order
        best_seasonal_order = temp_model.seasonal_order
        best_model = temp_model

print(f"Best seasonal frequency (m): {best_m}")
print(f"Best RMSE: {best_rmse}")
print(f"Best ARIMA Order: {best_order}")
print(f"Best Seasonal Order: {best_seasonal_order}")
```

```python
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Custom wrapper for ARIMA model to use with scikit-learn's GridSearchCV
class ARIMAWrapper(BaseEstimator, RegressorMixin):
    def __init__(self, order=(1, 0, 1), seasonal_order=(0, 0, 0, 0)):
        self.order = order
        self.seasonal_order = seasonal_order
        self.model = None

    def fit(self, X, y):
        self.model = ARIMA(endog=y, order=self.order, seasonal_order=self.seasonal_order, enforce_stationarity=False)
        self.model_fit = self.model.fit()
        return self

    def predict(self, X):
        return self.model_fit.forecast(steps=len(X))
```

```python
# Create a smaller parameter grid
param_grid = {
    'order': [(p, 0, q) for p in range(0, 2) for q in range(0, 2)],
    'seasonal_order': [(P, D, Q, S) for P in range(0, 2) for D in range(0, 2) for Q in range(0, 2) for S in [0, 7]]
}

# Initialize the GridSearchCV
grid = GridSearchCV(estimator=ARIMAWrapper(), param_grid=param_grid, scoring='neg_mean_squared_error', cv=3, verbose=1, n_jobs=-1)

# Fit to the data
grid.fit(train.index, train['PRCP'])

# Summary of the best model
print(f'Best ARIMA{grid.best_params_["order"]}x{grid.best_params_["seasonal_order"]} model - MSE:{-grid.best_score_}')

# Predicting and calculating MSE
y_pred = grid.best_estimator_.predict(test.index)
mae = mean_absolute_error(test['PRCP'], y_pred)
rmse = sqrt(mean_squared_error(test['PRCP'], y_pred))

print(f'Mean Absolute Error (MSE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```
